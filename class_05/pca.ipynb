```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
ds_test = pd.read_csv('train.csv')
# ds_test = pd.read_csv('test.csv')
```

```python
data = ds_test.values[:, 1:]
labels = ds_test.values[:, 0]
print labels.shape
print data.shape
# means = np.mean(data[:, :], axis=0)
# cm = np.cov(data[:, :], rowvar=0)

data_std = (data - data.mean(axis=0))/(data.std(axis=0)+1e-03)
print data_std.shape

mean_std = np.mean(data_std, axis=0)
cm_std = np.cov(data_std, rowvar=0)
```

```python
# print mean_std[:10]
# print cm_std[120:125, 120:125]
print cm_std[400:405, 400:405]
```

```python
eig_val_std, eig_vec_std = np.linalg.eig(cm_std)
```

```python
print eig_val_std[:5]
```

```python
plt.figure(0)
plt.plot(eig_val_std[:])
plt.show()
```

```python
all_vals = []
for (ix, iv) in zip(eig_val_std, eig_vec_std):
    all_vals.append([ix, iv])

print len(all_vals)
print all_vals[130][0]
```

```python
new_vals = sorted(all_vals, key=lambda x: x[0], reverse=True)
print new_vals[130][0]

eig_val = []
eig_vec = []

for ix in new_vals:
    eig_val.append(ix[0])
    eig_vec.append(ix[1])

eig_val = np.asarray(eig_val)
eig_vec = np.asarray(eig_vec)

print eig_val.shape, eig_vec.shape
```

```python
eig_cm = np.cumsum(eig_val)/np.sum(eig_val)
print eig_cm.shape
print eig_cm[:10]
```

```python
plt.figure(0)
plt.plot(eig_cm)
plt.show()
```

```python
eig_pairs = [(np.abs(eig_val_std[i]), eig_vec_std[:,i])
             for i in range(len(eig_val_std))]
eig_pairs.sort(key=lambda x:x[0])
eig_pairs.reverse()
```

```python
vals = np.zeros(len(eig_pairs))
for ix in range(len(eig_pairs)):
    vals[ix] = eig_pairs[ix][0]
print vals[700]
plt.plot(vals[:70])
plt.show()
```

```python
eig_cm = np.cumsum(vals)/np.sum(vals)


for ix in range(eig_cm.shape[0]):
    if eig_cm[ix] >= 0.99:
        print ix
        break


plt.plot(eig_cm)
plt.show()
```

```python
trans_cm = np.cov(X, rowvar=0)
print trans_cm[:4, :4]
new_mat = np.asarray(trans_cm, dtype='uint8')
print new_mat[:10, :10]
```

```python
v = pca.explained_variance_ratio_
cs = np.cumsum(v)/np.sum(v)
```
